```python
# Import the required libraries
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder 
```

```python
# import the dataset 
df=pd.read_csv('GroceryData.csv')
display(df)
```

```python
# we do not need customer ID. So drop it.
df.drop(columns='CustomerID',inplace=True)
```

```python
# calculate minimum support
freq=apriori(df,min_support=0.5,use_colnames=True)
freq
```

```python
association_rules(freq,metric='confidence',min_threshold=0.25)
```

```python
df=pd.read_csv('bread basket.csv')
df.head()
```

```python
df=df.groupby('Transaction')['Item'].apply(list).reset_index(name='Item List')
df.head()
```

```python
dummies=TransactionEncoder()
dummi=dummies.fit(df['Item List']).transform(df['Item List'])
```

```python
dummi=pd.DataFrame(dummi,columns=dummies.columns_)
```

```python
dummi
```

```python
freq=apriori(dummi,min_support=0.01,use_colnames=True)
freq
```

```python
association_rules(freq,metric='confidence',min_threshold=0.5)
```
